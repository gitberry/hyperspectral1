```python
#iterate through all channels
import matplotlib.pyplot as plt
import numpy as np
import cv2
#delayed attempt to finesse the plot colors -> colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'w']

#im = cv2.imread('../NesbitTry1/SMALL.png')
im = cv2.imread('../NesbitTry1/NesbitRGBSat.png')
print(im.shape)
#cv2.imshow('dst',im)
#print(im)
width,height,depth=im.shape
print("height:", str(height)," width:", str(width)," depth:",str(depth))
zX = np.empty(height) 
zR = np.empty(height) 
zG = np.empty(height) 
zB = np.empty(height) 
#print ("zX shape:",zX.shape)

a1 = ""
TheIncrement = 0
for yrange in range(height):
    zX[yrange] = TheIncrement
    zR[yrange] = 0
    zG[yrange] = 0
    zB[yrange] = 0
    at = ""
    for xrange in range(width):        
        for zrange in range(depth):
            at = at + "[" + str(yrange) + "," + str(xrange) + "," + str(zrange) + ":" + str(im[yrange,xrange,zrange]) + "]" 
            if zrange == 0 : zR[yrange] = zR[yrange] + im[yrange,xrange,zrange]
            if zrange == 1 : zG[yrange] = zG[yrange] + im[yrange,xrange,zrange]
            if zrange == 2 : zB[yrange] = zB[yrange] + im[yrange,xrange,zrange]            
            at = at + " {" + str(zR[yrange])+","+str(zG[yrange])+","+str(zB[yrange])+"} "
    #print(at)
    TheIncrement = TheIncrement + 1
    # We turn the sum into a arithmetic mean - aka average to see what our average pixel  RGB values are for that column... 
    zR[yrange] = zR[yrange]/width # + im[yrange,xrange,zrange]
    zG[yrange] = zG[yrange]/width # + im[yrange,xrange,zrange]
    zB[yrange] = zB[yrange]/width # + im[yrange,xrange,zrange]  
#thisPlt = plt.plot(zX, zR, zG, zB)
#thisPlt.annotate(f"({r}, {g}, {b})")
#thisPlt
plt.subplot(1,2,1)
plt.plot(zX, zR, color='r')
plt.plot(zX, zG, color='g')
plt.plot(zX, zB, color='b')
plt.title("Orig")
plt.xlabel("west-east")
plt.ylabel("avg val")

im = cv2.imread('../NesbitTry1/NesbitRGBSatBLUESTRIP.png')
print(im.shape)
#cv2.imshow('dst',im)
#print(im)
width,height,depth=im.shape
#height,width,depth=im.shape
print("height:", str(height)," width:", str(width)," depth:",str(depth))
zX = np.empty(height) 
zR = np.empty(height) 
zG = np.empty(height) 
zB = np.empty(height) 
#print ("zX shape:",zX.shape)

a1 = ""
TheIncrement = 0
for yrange in range(height):
    zX[yrange] = TheIncrement
    zR[yrange] = 0
    zG[yrange] = 0
    zB[yrange] = 0
    at = ""
    for xrange in range(width):        
        for zrange in range(depth):
            at = at + "[" + str(yrange) + "," + str(xrange) + "," + str(zrange) + ":" + str(im[yrange,xrange,zrange]) + "]" 
            if zrange == 0 : zR[yrange] = zR[yrange] + im[yrange,xrange,zrange]
            if zrange == 1 : zG[yrange] = zG[yrange] + im[yrange,xrange,zrange]
            if zrange == 2 : zB[yrange] = zB[yrange] + im[yrange,xrange,zrange]            
            at = at + " {" + str(zR[yrange])+","+str(zG[yrange])+","+str(zB[yrange])+"} "
    #print(at)
    TheIncrement = TheIncrement + 1
    # We turn the sum into a arithmetic mean - aka average to see what our average pixel  RGB values are for that column... 
    zR[yrange] = zR[yrange]/width # + im[yrange,xrange,zrange]
    zG[yrange] = zG[yrange]/width # + im[yrange,xrange,zrange]
    zB[yrange] = zB[yrange]/width # + im[yrange,xrange,zrange]  
#thisPlt2 = plt.plot(zX, zR, zG, zB)
#thisPlt.annotate(f"({r}, {g}, {b})")
#thisPlt
plt.subplot(1,2,2)
plt.plot(zX, zR, color='r')
plt.plot(zX, zG, color='g')
plt.plot(zX, zB, color='b')
plt.title("BLUE stripe")
plt.xlabel("west-east")
plt.ylabel("avg val")
```
